```python
url = "https://www.marca.com/en/football/uefa-euro/winners.html"

page = requests.get(url).text
page
```

```python
soup = BeautifulSoup(page, 'html5lib')
soup.prettify()
```

```python
finalists = soup.table
```

```python
finalists
```

```python
column_titles = soup.table.find(class_="ue-table-record__thead").text.replace("\t","")
column_titles = column_titles.replace('\n'," ").split(" ")
column_titles
```

```python
i = 0
for title in column_titles:
    if title == '':
        column_titles.pop(i)
    i = i + 1
```

```python
column_titles
```

```python
finalists = soup.table.find(class_="ue-table-record__tbody")
finalists
```

```python
year_data = finalists.find(class_="ue-table-record__year").text
year_data
```

```python
host_data = finalists.find(class_="ue-table-record__sede").text
host_data
```

```python
winner_data = finalists.find(class_="ue-table-record__name is-marked").text
winner_data
```

```python
runner_up_data = finalists.find(class_="ue-table-record__td",scope="row")
runner_up_data = runner_up_data.find_next_sibling().text
runner_up_data
```

```python
finalists = soup.table.find(class_="ue-table-record__tbody")
f1 = finalists.find_all(class_="ue-table-record__tr")

f1

year = []
host = []
winner = []
runner_up = []

for finalist in f1:
    year_data = finalist.find(class_="ue-table-record__year").text
    host_data = finalist.find(class_="ue-table-record__sede").text
    winner_data = finalist.find(class_="ue-table-record__name is-marked").text
    runner_up_data = finalist.find(class_="ue-table-record__td",scope="row")
    runner_up_data = runner_up_data.find_next_sibling().text
    year.append(year_data)
    host.append(host_data)
    winner.append(winner_data)
    runner_up.append(runner_up_data)
```

```python
import pandas as pd

df = pd.DataFrame(list(zip(year,host,winner,runner_up)),columns=column_titles)
df
```
