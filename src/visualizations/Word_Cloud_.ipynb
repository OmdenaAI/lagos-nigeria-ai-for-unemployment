```python
import numpy as np
import pandas as pd
import re
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

import matplotlib.pyplot as plt
% matplotlib inline
```

```python
df = pd.read_csv("/content/drive/MyDrive/OMDENA INC/jobb_data.csv",index_col=0)
df.head()
```

```python
df.isna().sum()
```

```python
df.dropna(inplace = True)
df
```

```python
text = " ".join(Location for Location in df.Location)
text
```

```python
word_cloud = WordCloud(collocations = False, background_color = 'white').generate(text)
```

```python
plt.imshow(word_cloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
