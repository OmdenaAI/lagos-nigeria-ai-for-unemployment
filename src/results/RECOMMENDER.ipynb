# Word Cloud

```python
import matplotlib.pyplot as plt
import re
from wordcloud import WordCloud
```

```python
def clean_text(text):
    #Remove RT
    text = re.sub(r'RT', '', text)
    
    #Fix &
    text = re.sub(r'&amp;', '&', text)
    
    #Remove punctuations
    text = re.sub(r'[?!.;&/:,#@-]', '', text)
  

    #Convert to lowercase to maintain consistency
    text = text.lower()
    return text

text = df_copy["combined_corpus"].apply(lambda x: clean_text(x))
```

```python
def gen_freq(text):
    #Will store the list of words
    word_list = []

    #Loop over all the tweets and extract words into word_list
    for tw_words in text.split():
        word_list.extend(tw_words)

    #Create word frequencies using word_list
    #print(word_list)
    word_freq = pd.Series(word_list).value_counts()

    #Print top 20 words
    word_freq[:20]
    
    return word_freq

freq = gen_freq(text.str)
freq
```

```python
np.random.seed(43)
real_words = freq.drop(labels= stopwords, errors= 'ignore')
wcc = WordCloud(width=550, height=330, max_words=50, background_color="white", colormap = "plasma").generate_from_frequencies(real_words[:50])
plt.figure(figsize=(12, 14))
plt.imshow(wcc, interpolation='bilinear')
plt.title('Word Cloud of Top 50 words in Jobs Dataset', fontsize = 20)
plt.axis('off')
plt.show()
```

```python
# nearest neighbors set to return the top 5 recommendations

nbrs =  NearestNeighbors(n_neighbors=5).fit(Array_DataFrame)
```

```python
# function to carry out the nearest neighbors algorithm
# and return a dataframe showing the distance calculated and their respective index from the data.

def n_neighbors(data, test):
    distance, indicies = nbrs.kneighbors(test, return_distance=True)
    result_df = pd.DataFrame(distance[0], index=indicies[0], columns=['distance'])
    return result_df
```

## Cosine similarity algorithm

```python
# importing the cosine similarity algo
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# getting the first person from the processed cv dataset
test = cv_data.loc[[0],:]
test
```

```python
# function to carry out the nearest neighbors algorithm
# and return a dataframe showing the cosine similarity calculated and their respective index from the data.

def cosine(data, test):
    result = cosine_similarity(data, test)
    result_df = pd.DataFrame(result, columns=['distance'])
    result_df.sort_values('distance', ascending=False, inplace=True)
    return result_df.head(5)
```

```python
trial2 = cosine(Array_DataFrame, test)
trial2
```

```python
for index in trial2.index:
    print(df.loc[index,:])
    print('')
```
